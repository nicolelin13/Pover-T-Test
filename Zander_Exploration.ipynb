```python
def load_files():
    temp_dict = {}
    temp_file = None
    # Iterate over directories
    for directory in ['train', 'test']:
        # Iterate over files
        for file in os.listdir(directory):
            # Ignore anything that isn't a file
            if os.path.isfile(os.path.join(train_dir, file)):
                # Split files into base_name and extension
                base_name, extension = os.path.splitext(file)
                # Ignore anything that isn't a .csv file
                if extension == '.csv':
                    # Split base_name into component parts
                    (country, rollup, split) = base_name.split('_')
                    # Ensure temp_dict has necessary keys
                    if split not in temp_dict.keys():
                        temp_dict[split] = {}
                    if country not in temp_dict[split].keys():
                        temp_dict[split][country] = {}
                    # Read data into temporary dictionary
                    temp_dict[split][country][rollup] = pd.read_csv(os.path.join(train_dir, file))
    return temp_dict
```
